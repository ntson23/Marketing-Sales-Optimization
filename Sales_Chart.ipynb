```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Database - Q3_2020.xlsx', 'Sales')
```

```python
df.info()
```

```python
df[df.duplicated(keep=False)]
```

```python
df.drop_duplicates(inplace = True)
```

```python
print(df['Tổng tiền'].isna().sum()) #không có gt trống nhưng nhiều hơn các cột khác 1 gt. Nghi có nhiều cột NaN
```

```python
df[df.isnull().any(axis=1)] #hiển thị tất cả các dòng có ít nhất một giá trị NaN ở bất kỳ cột nào
```

```python
error_row = df[((df['Tổng tiền'] == 0) & df[['Giờ', 'Khách hàng', 'SĐT']].isnull().all(axis=1))]
# Trong pandas, axis được sử dụng để xác định hướng của thao tác trên DataFrame:
# axis=0 → Áp dụng theo hàng (row) (từ trên xuống dưới).
# axis=1 → Áp dụng theo cột (column) (từ trái sang phải).
print(error_row)
```

```python
df = df.drop(error_row.index) #xóa cột bị sai
```

```python
df['Type of Lead'] = df['Type of Lead'].replace("Con đang học lớp 7", "dathang")
df['Type of Lead'] = df['Type of Lead'].fillna('dathang')
#xử lý cột Type of lead cho cùng định dạng
```

```python
df['Số tiền giảm giá'] = df['Số tiền giảm giá'].astype(str).str.replace("-", "").astype(float)
```

```python
print(df['Close Date'].dtype)
```

```python
df.loc[df['Close Date'].dt.year == 2023, 'Close Date'] -= pd.DateOffset(years = 3)
```
