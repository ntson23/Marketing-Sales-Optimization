```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Database - Q3_2020.xlsx', 'Vận đơn')
```

```python
df.info()
```

```python
df = df.drop(columns=['Lý do hủy đơn'])
```

```python
df = df.drop(columns=['Ngày hẹn giao', 'Ghi chú sản phẩm', 'Serial' ])
```

```python
#tag đơn hàng null nhiều và không có giá trị phân tích => xóa
df = df.drop(columns=['Tags đơn hàng'])
```

```python
df = df.drop(columns=['Ghi chú đơn hàng'])
```

```python
error_row = df[((df['Mã sản phẩm'].notna()) & df[['Mã đơn hàng', 'Mã vận đơn']].isna().all(axis=1))]
```

```python
df = df.drop(error_row.index)
```

```python
status_customer = df['Trạng thái đối tác'].isnull()
```

```python
print(status_customer)
```

```python
pd.set_option('display.max_columns', None)
Delivery_date = df[df['Ngày giao hàng'].isnull()]
print(Delivery_date)
```

```python
df['SĐT người nhận'] = (df['SĐT người nhận']
    .str.replace(" ", '', regex=True)
    .str.replace("^84", "", regex=True)
    .astype('int', errors='ignore')
)
```

```python
df.to_excel('Vận đơn.xlsx')
```
