```python
nan_row = df[pd.to_numeric(df['ME1 (ME đã bao gồm thuế + phí)'], errors='coerce').isnull()] #lấy ra những giá trị không phải là số ở cột ME1
```

```python
# Giả sử cột chi phí MKT có tên là 'Chi phí MKT'
nan_row.loc[nan_row['Chi phí Marketing'] > 100000, 'ME1 (ME đã bao gồm thuế + phí)'] *= 1.08
```

```python
nan_row
```

```python
df.loc[nan_row.index, 'ME1 (ME đã bao gồm thuế + phí)'] = 0
```

```python
print(df.loc[nan_row.index, 'ME1 (ME đã bao gồm thuế + phí)'])
```

```python
df.loc[nan_row.index, 'ME1 (ME đã bao gồm thuế + phí)'] = nan_row['Chi phí Marketing'].apply(
    lambda x: x * 1.08 if x > 100000 else x
)
```

Làm sạch xong bắt đầu vẽ chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print( df['Date'].dtypes)
```

```python
# Nhóm dữ liệu theo ngày và tính tổng chi phí marketing
df_grouped = df.groupby("Date")["Chi phí Marketing"].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_grouped, x="Date", y="Chi phí Marketing", color="blue")
```

```python
plt.xticks(rotation=45)
plt.xlabel("Ngày")
plt.ylabel("Chi phí Marketing")
plt.title("Chi phí Marketing theo ngày")
```

```python
plt.show()
```

```python
print(df_grouped.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(df_grouped["Date"], df_grouped["Chi phí Marketing"], marker='o', linestyle='-')
plt.xticks(rotation=45)
plt.xlabel("Ngày")
plt.ylabel("Chi phí Marketing")
plt.title("Chi phí Marketing theo ngày")
plt.show()
```
